```python
data = {
    'reservation_status_date': [
        '2016-01-01', '2016-05-01', '2016-09-01',
        '2017-01-01', '2017-05-01', '2017-09-01'
    ],
    'adr': [100, 120, 130, 90, 110, 105]
}

df = pd.DataFrame(data)

# Convert the date column to datetime format
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Filter data for the specified date range
start_date = pd.to_datetime('2016-01-01')
end_date = pd.to_datetime('2017-09-01')
filtered_df = df[(df['reservation_status_date'] >= start_date) & (df['reservation_status_date'] < end_date)]

# Create the line plot
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(filtered_df['reservation_status_date'], filtered_df['adr'], label='ADR')
plt.xlabel('Date')
plt.ylabel('ADR')
plt.legend(fontsize=12)
plt.xticks(rotation=45)  
plt.show()
```
